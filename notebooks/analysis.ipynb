```python
import pandas as pd
df = pd.read_csv(r"C:\Users\LENOVO\Documents\Network log Analysis\data\UNSW_NB15_training-set.csv")

print(df.columns)
```

```python
# these are the columns I suggest to keep
keep_cols = [
    'dur', 'proto', 'service', 'state', 
    'spkts', 'dpkts', 'sbytes', 'dbytes', 
    'sload', 'dload', 'attack_cat', 'label'
]

# create a cleaned dataframe
df_clean = df[keep_cols]
df_clean.head()

```

```python
import matplotlib.font_manager
matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

plt.rcParams['font.family'] = 'serif'   # or 'sans-serif'
plt.rcParams['font.serif'] = ['Cambria']   # or any installed serif font
plt.rcParams['font.size'] = 12   # overall font size

# counts of normal vs attack
df_clean['label'].value_counts().plot(kind='bar')
plt.xticks([0, 1], ['Normal', 'Attack'])
plt.title("Normal vs Attack Traffic")
plt.ylabel("Number of Records")
plt.show()
```

This plot shows how many flows are normal vs how many are attacks. This is critical for understanding dataset imbalance.

```python

df_clean['attack_cat'] = df_clean['attack_cat'].replace('-', 'Normal')
df_clean['attack_cat'] = df_clean['attack_cat'].fillna('Normal')
# get the cleaned counts
attack_counts = df_clean['attack_cat'].value_counts()

colors = sns.color_palette("tab10", n_colors=len(attack_counts))
plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=attack_counts.values,
    y=attack_counts.index,
    palette="rocket",
   
)

plt.title("Distribution of Attack Categories", fontsize=16)
plt.xlabel("Number of Records", fontsize=14)
plt.ylabel("Attack Category", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# add labels on bars
for i, v in enumerate(attack_counts.values):
    ax.text(v + 500, i, str(v), va='center', fontsize=12)

plt.tight_layout()
plt.show()

```

Here we analyze which attack types are most frequent in the dataset.

```python
# get top protocols
proto_counts = df_clean['proto'].value_counts()
proto_counts = proto_counts[proto_counts > 1000].head(10)

# build a softened rocket palette
colors = sns.color_palette("rocket", n_colors=len(proto_counts), desat=0.6)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    y=proto_counts.index,
    x=proto_counts.values,
    palette=colors,
    
)

plt.title("Top Protocols with Significant Traffic", fontsize=16)
plt.xlabel("Number of Records", fontsize=14)
plt.ylabel("Protocol", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# add data labels on bars
for i, v in enumerate(proto_counts.values):
    ax.text(v + 500, i, str(v), va='center', fontsize=12)

plt.tight_layout()
plt.show()

```

```python
service_counts = df_clean[df_clean['service'] != '-']['service'].value_counts()
# keep only those above, say, 1000 flows
service_counts = service_counts[service_counts > 1000].head(10)

service_counts.plot(kind='bar', color='skyblue', figsize=(10,6),  edgecolor='black')
plt.title("Top Services with Significant Traffic", fontsize=16)
plt.ylabel("Number of Records", fontsize=14)
plt.xlabel("Service", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

Note: Flows with unknown service (â€˜-â€™) were excluded from this plot to focus on identified services.

```python
df_clean[['dur', 'sbytes', 'dbytes']].hist(figsize=(12,8), bins=50, log=True, color='skyblue')
plt.suptitle("Log-scaled distribution of numeric features")
plt.show()
```

Due to high skewness in network flow data, we applied log-scaling to visualize numeric features more effectively.

```python
corr = df_clean.corr(numeric_only=True)

plt.figure(figsize=(12,10))
sns.heatmap(
    corr, 
    cmap="viridis", 
    annot=True, 
    fmt=".2f",
    square=True, 
    linewidths=0.5,
    cbar_kws={"shrink":0.8}
)
plt.title("Correlation Heatmap", fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

```
